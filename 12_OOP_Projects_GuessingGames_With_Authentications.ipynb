```python
import random
class Register:
    def __init__(self):
        self.email = ''
        self.name = ''
        self.password = ''
        self.first_menu()
    
    def first_menu(self):
        userinput = input("""
        1. Press 1 to Register 
        2. Press 2 to Login
        3. Press 3 to Closed
        """) 
        
        if userinput == '1':
            self.register()
        if userinput == '2':
            self.login()
        else:
            exit()
            
    def register(self):
        r_email = input('Enter Email : ')
        r_name = input('Enter Name : ')
        r_password = input('Create Password : ')
        
        if r_email == self.email:
            print('Email Already Exists !')
        else:
            self.email = r_email
            self.name = r_name
            self.password = r_password
            print('Registrations Sucessfully Now login')
            print(self.email,self.name,self.password)
            self.first_menu() 
            
    def login(self):
        l_email = input('Enter email : ')
        l_password = input('Enter Password : ')
        if l_email == self.email:
            if l_password == self.password:
                print('Login Sucessfully')
            else:
                print('Wrong Password')
                self.login()
        else:
            print('Email Not Exists ! ')
            self.first_menu()
            
            
            
class Games(Register):
    def __init__(self):
        super().__init__()
        user_per = input('Press 0 To start a Games')
        if user_per == '0':
            self.random_number_guess()
        else:
            print('Invalid Input')
        
    
    def random_number_guess(self):
        ran = random.randint(0,100)
        user_num = int(input('Guess the number'))
        counter = 0
        
        while user_num != ran:
            if user_num > ran:
                user_num = int(input('Guess Lower'))
            elif user_num < ran:
                user_num = int(input('Guess Higher'))
                
            counter += 1
            
                  
        print(f'Congratulations {self.name} Your guess is',user_num)
        print('Game Over')
        print('Your Total Attempt for guessing number is = ', counter)
        user_want = input('Play Again press Y/N')
        if user_want == 'Y':
            self.random_number_guess()
        else:
            print('Programme Close')
        
```
